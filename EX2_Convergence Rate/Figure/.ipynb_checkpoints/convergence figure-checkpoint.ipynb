```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

```python
file_name = "EX2_results_all.xlsx"
sheet_names = ["10d_5q", "10d_25q", "50d_5q"]  # 三个 sheet 的顺序 = 三列顺序
```

```python
def slope_minus_one(y0, gamma_arr):
    g0 = gamma_arr[0]
    return y0 * (gamma_arr /g0) ** (-1)
```

```python
sns.set(style="white", context="talk")
plt.rcParams["axes.grid"] = False

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Times"],
    "text.latex.preamble": r"\usepackage{newtxtext}\usepackage{newtxmath}",
    "xtick.labelsize": 15,
    "ytick.labelsize": 15
})


color_LR   = "tab:orange"
color_KRR  = "tab:blue"
color_kNN  = "tab:green"
color_KS   = "tab:purple"
color_slope = "gray"
```

```python
cfg = {
    "10d_5q": {
        "x_lim":        (11000, 41000),    
        "x_ticks":      np.array([12000, 18000, 26000, 40000]),
        "x_labels":     ["12", "18", "26", "40"],
        "x_label_LR":   r"$\Gamma\ (\times 10^3)$",
        "x_label_KS":   r"$\Gamma\ (\times 10^3)$",
        "lr_ylim":      (1.8e-4, 2.75e-3),
        "lr_yticks":    np.array([2.5e-3, 1e-3, 0.5e-3, 0.2e-3]),
        "lr_ytlabels":  ["2.5", "1.0", "0.5", "0.2"],
        "ks_ylim":      (0.18, 2.0),
        "ks_yticks":    np.array([1.8, 0.8, 0.4, 0.2]),
        "ks_ytlabels":  ["1.8", "0.8", "0.4", "0.2"],
    },
    "10d_25q": {
        "x_lim":        (11000, 41000), 
        "x_ticks":      np.array([12000, 18000, 26000, 40000]),
        "x_labels":     ["12", "18", "26", "40"],
        "x_label_LR":   r"$\Gamma\ (\times 10^3)$",
        "x_label_KS":   r"$\Gamma\ (\times 10^3)$",
        "lr_ylim":      (1.8e-4, 2.75e-3),
        "lr_yticks":    np.array([2.5e-3, 1e-3, 0.5e-3, 0.2e-3]),
        "lr_ytlabels":  ["2.5", "1.0", "0.5", "0.2"],
        "ks_ylim":      (0.18, 2.0),
        "ks_yticks":    np.array([1.8, 0.8, 0.4, 0.2]),
        "ks_ytlabels":  ["1.8", "0.8", "0.4", "0.2"],
    },
    "50d_5q": {
        "x_lim":        (28500, 78000),  
        "x_ticks":      np.array([30000, 40000, 55000, 75000]),
        "x_labels":     ["30", "40", "55", "75"],
        "x_label_LR":   r"$\Gamma\ (\times 10^4)$",
        "x_label_KS":   r"$\Gamma\ (\times 10^3)$",
        "lr_ylim":      (6.5e-4, 1.2e-2),
        "lr_yticks":    np.array([0.7e-3, 2e-3, 5e-3, 10e-3]),
        "lr_ytlabels":  ["0.7", "2", "5", "10"],
        "ks_ylim":      (0.099, 1.4),
        "ks_yticks":    np.array([1.2, 0.5, 0.2, 0.1]),
        "ks_ytlabels":  ["1.2", "0.5", "0.2", "0.1"],
    }
}
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))  # 2 行 3 列

for col, sheet_name in enumerate(sheet_names):
    df = pd.read_excel(file_name, sheet_name=sheet_name)
    df = df.dropna(subset=["Gamma", "LR", "KRR", "kNN", "KS"])
    
    gamma = df["Gamma"].to_numpy()

    
    ref_LR = slope_minus_one(df["LR"].iloc[0], gamma)
    ref_KS = slope_minus_one(df["KS"].iloc[0], gamma)
    
    cfg_s = cfg[sheet_name]
    
    
    ax_lr = axes[0, col]
    ax_lr.loglog(gamma, df["LR"],  marker="o", linestyle="-", label="LR",  color=color_LR)
    ax_lr.loglog(gamma, df["KRR"], marker="s", linestyle="-", label="KRR", color=color_KRR)
    ax_lr.loglog(gamma, ref_LR, linestyle="--", color=color_slope, label="slope = -1")


    ax_lr.set_xlim(*cfg_s["x_lim"])
    ax_lr.xaxis.set_major_locator(mticker.FixedLocator(cfg_s["x_ticks"]))
    ax_lr.set_xticklabels(cfg_s["x_labels"])
    ax_lr.xaxis.set_minor_locator(mticker.NullLocator())    # ax_lr.set_xlabel(cfg_s["x_label_LR"])
    
    ax_lr.set_ylim(*cfg_s["lr_ylim"])
    ax_lr.set_yticks(cfg_s["lr_yticks"])
    ax_lr.set_yticklabels(cfg_s["lr_ytlabels"])
    if col == 0:
        ax_lr.set_ylabel(r" Optimality Gap  $(\times 10^{-3})$")
    ax_lr.legend(loc="lower left", fontsize =14,framealpha=0.3)
    
    title_map = {"10d_5q":"$d=10,\ q=5$", "10d_25q":"$d=10,\ q=25$", "50d_5q":"$d=50,\ q=5$"}
    ax_lr.set_title(title_map[sheet_name])
    
    ax_ks = axes[1, col]
    ax_ks.loglog(gamma, df["kNN"], marker="^", linestyle="-", label="kNN", color=color_kNN)
    ax_ks.loglog(gamma, df["KS"],  marker="d", linestyle="-", label="KS",  color=color_KS)
    ax_ks.loglog(gamma, ref_KS, linestyle="--", color=color_slope, label="slope = -1")
    
    ax_ks.set_xlim(*cfg_s["x_lim"])
    ax_ks.xaxis.set_major_locator(mticker.FixedLocator(cfg_s["x_ticks"]))
    ax_ks.set_xticklabels(cfg_s["x_labels"])
    ax_ks.xaxis.set_minor_locator(mticker.NullLocator())

    
    ax_ks.set_ylim(*cfg_s["ks_ylim"])
    ax_ks.set_yticks(cfg_s["ks_yticks"])
    ax_ks.set_yticklabels(cfg_s["ks_ytlabels"])
    if col == 0:
        ax_ks.set_ylabel("Optimality Gap")
    ax_ks.legend(loc="lower left", fontsize =14, framealpha=0.3)

plt.tight_layout()
plt.savefig("convergence.pdf",dpi=600,bbox_inches = 'tight')

plt.show()
```
